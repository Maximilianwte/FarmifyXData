```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_excel("Data/TestDataGeo.xlsx")
```

```python
df.head()
```

```python
def grouper(df):
    groups = list()
    notX = 0
    for yCOR in range(114, 155, 8):
        for xCOR in range(-22, -43, -4):
            xGroup = list()
            yGroup = list()
            group = {"id" : int(len(groups))+1, "location": "", "farmsIncluded" : []}
            for i in range(0, len(df)):
                dec = df.GeoLocation[i].find(",")
                xFarm = float(df.GeoLocation[i][0:dec])
                yFarm = float(df.GeoLocation[i][dec+1:])
                if (xFarm >= xCOR) and (xFarm < xCOR + 8) \
                and (yFarm >= yCOR) and (yFarm < yCOR + 4):
                    group["farmsIncluded"].append(df.Index[i])
                    xGroup.append(xFarm)
                    yGroup.append(yFarm)
            
            if len(group["farmsIncluded"]) > 0:  
                group["location"] = f"{np.mean(xGroup)},{np.mean(yGroup)}"
                groups.append(group)
            else:
                notX = notX + 1
    print(f"Number of groups without data: {notX}")
    df = pd.DataFrame(groups)
    return df
```

```python
df1 = grouper(df)
```
