```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_excel("Data/TestData.xlsx")
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.head()
```

```python
def grouper(df):
    groups = list()
    for yCOR in range(114, 155, 8):
        for xCOR in range(-22, -43, -4):
            xGroup = list()
            yGroup = list()
            group = {"id" : int(len(groups))+1, "location": "", "numberInGroup" : 0, "farmsIncluded" : []}
            for i in range(0, len(df)):
                xFarm = int(df.GeoLocation[i][:3])
                yFarm = int(df.GeoLocation[i][12:15])
                if (xFarm >= xCOR) and (xFarm < xCOR + 8) \
                and (yFarm >= yCOR) and (yFarm < yCOR + 4):
                    group["numberInGroup"] = group["numberInGroup"] + 1
                    group["farmsIncluded"].append(df.index[i])
                    xGroup.append(xFarm)
                    yGroup.append(yFarm)
            
            if group["numberInGroup"] > 0:  
                group["location"] = f"{np.mean(xGroup)},{np.mean(yGroup)}"
                groups.append(group)
    df = pd.DataFrame(groups)
    return df
```

```python
df = grouper(df)
```
